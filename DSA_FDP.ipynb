## **MODEL** ##

### Label encoding ###

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df1['AIRLINE']=label.fit_transform(df1['AIRLINE'])
df1['TAIL_NUMBER']=label.fit_transform(df1['TAIL_NUMBER'])
df1['ORIGIN_AIRPORT']=label.fit_transform(df1['ORIGIN_AIRPORT'])
df1['DESTINATION_AIRPORT']=label.fit_transform(df1['DESTINATION_AIRPORT'])
df1.head(5)
```

### Droping unnecessary data ###

```python
df1=df1.drop(['YEAR','CANCELLATION_REASON','AIR_SYSTEM_DELAY','SECURITY_DELAY','AIRLINE_DELAY','LATE_AIRCRAFT_DELAY','WEATHER_DELAY'],axis=1)
df1.head()
```

### Spliting x and y ###

```python
x=df1.drop('ARRIVAL_DELAY',axis=1)
y=df1['ARRIVAL_DELAY']
#x.head()
#y.head()
```

### Training and testing percentage ### 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,test_size=0.3)
```

### Model formation ###

```python
from sklearn.linear_model import LinearRegression
linear=LinearRegression()

#training
fit=linear.fit(x_train,y_train)

#testing
pred=linear.predict(x_test)
```

### Training and Testing score ###

```python
linear.score(x_test,y_test)
```

```python
linear.score(x_train,y_train)
```
